```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def z_score(x,u,std):
    return (x - u) / std
```

```python
def variance(x):
    mean = 0
    sq_mean = 0
    N = len(x)
    for i in range(N):
        mean = mean + x[i]
        sq_mean = sq_mean + x[i]**2
    return (1/N)*sq_mean - ((1/N)*mean)**2
def std(x):
    return np.sqrt(variance(x))
```

```python
def normal(x):
    mean = np.mean(x)
    std1 = std(x)
    return (1/(np.sqrt(np.pi*2)*std1))*np.exp((-1/2)*((z_score(x,mean,std1))**2))
```

```python
def plot_hist(data_1):
    fig, ax = plt.subplots(figsize=(10,7),tight_layout = True)
    ax.grid(b = True,
            linestyle = '-',
            color='aqua',
            alpha = 0.2)
    for s in ['top','bottom','left','right']:
        ax.spines[s].set_visible(False)
    ax.hist(
        data_1,
        color = 'lime',
        bins = 100
    )
    plt.xlabel("data")
    plt.ylabel("values")
    ax.xaxis.set_ticks_position("none")
    ax.yaxis.set_ticks_position("none")
```

```python
data = np.random.randint(0,100,500)
plot_hist(normal(data))
```

```python
import math
def binomial_prob(p_success,n,k):
     comb = (math.factorial(n))/((math.factorial(k))*math.factorial(n-k))
     return comb*(p_success**k)*((1-p_success)**(n - k))
```

```python
print(binomial_prob(1/6,6,2))
```

```python
def mean_simlation(x,p):
    mean = 0
    for i in range(len(x)):
        mean = mean + x[i]*p[i]
    return mean
```

```python
def std_simulation(x,p):
    mean2 = mean_simlation(x,p)
    std2 = 0
    for i in range(len(x)):
        std2 = std2 + ((x[i] - mean2)**2)*p[i]
    return np.sqrt(std2)
```

```python
def normal_curve():
    x = [i for i in range(0,50,1)]
    z = normal(x)
    y = [np.mean(x)]*len(x)
    plt.plot(x, z , color = 'lime',label="curve")
    plt.plot(y,x,label="mean" ,color='grey')
    plt.xlabel("X")
    plt.ylabel("pdf(X)")
    plt.ylim(0.006,0.03)
    plt.legend()
    plt.show()
```
