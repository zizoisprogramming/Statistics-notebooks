```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
fig = plt.figure(figsize=(10,5))
names = ["ziad","samer",'mo']
vals = [4,5,4]
plt.bar(names,vals,color ='aqua')
plt.xlabel("names")
plt.ylabel("values")
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,5))
ax.barh(names,vals)
for s in ['top','bottom','left','right']:
    ax.spines[s].set_visible(False)
ax.grid(b =  True, color ='lime',
        linestyle ='-', linewidth = 0.3,
        alpha = 0.4)
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
plt.xlabel("vals")
```

```python
fruits = ['apple','banana','strawberry','grapes','mango']
values = [rd.randint(a= 1, b= 8) for i in range(len(fruits))]
plt.pie(values, labels=fruits,
        colors=['green','yellow','red','maroon','orange'],
            shadow=True,explode=[0.1 for i in range(len(fruits))],
                autopct='%1.1f%%')
plt.legend(bbox_to_anchor=(1,1), loc="best")
```

```python
data_1 = [rd.randint(a=1,b=100) for i in range(100)]
data_2 = [rd.randint(a=2,b=200) for i in range(100)]
data_3 = [rd.randint(a=3,b=300) for i in range(100)]
data = [data_1,data_2,data_3]
fig = plt.figure(figsize=(10,5))
ax= fig.add_axes([1,1,1,1])
bp = ax.boxplot(data)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,7),tight_layout = True)
ax.grid(b = True,
        linestyle = '-',
        color='aqua',
        alpha = 0.2)
for s in ['top','bottom','left','right']:
    ax.spines[s].set_visible(False)
ax.hist(
    data_1,
    color = 'lime',
    bins = 20
)
plt.xlabel("data")
plt.ylabel("values")
ax.xaxis.set_ticks_position("none")
ax.yaxis.set_ticks_position("none")
```
